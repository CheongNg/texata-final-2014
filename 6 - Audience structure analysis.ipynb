## Audience structure analysis

We are interseted in validating the assumption about there being "mainly askers" and "mainly answerers" on the forum.
Obviously, one way to do that is examining the question/answer stats of the users. For simplicity we'll only focus on the RS forum.

```python
import tx
from tx.db import *
db_url = 'postgresql://texata:__austin__@localhost:5432/texata_rs'
connect_db(db_url)
```

```python
%%time
from sqlalchemy import func
from collections import Counter
user_reply_count = Counter(dict(DBSession.query(Reply.user_id, func.count(Reply.id)).group_by(Reply.user_id).all()))
user_question_count = Counter(dict(DBSession.query(Post.user_id, func.count(Post.id)).group_by(Post.user_id).all()))
user_vote_count = Counter(dict(DBSession.query(Post.user_id, func.sum(Post.vote_count)).group_by(Post.user_id).all()))
from tx.db import Engine
replies = Engine.execute("select reply.user_id, count(reply.id) from reply left join post on reply.post_id = post.id where reply.user_id <> post.user_id group by reply.user_id")
user_help_count = Counter(dict(list(replies)))
```

```python
import numpy as np
user_ids = DBSession.query(User.id).all()
X = np.array([(uid, user_help_count[uid], user_question_count[uid], user_vote_count[uid]) for (uid,) in user_ids])
```

```python
%pylab inline
```

```python
scatter(X[:,1], X[:,2])
xlabel("Replies to others")
ylabel("Questions");
```

For one, this plot immediately tells us there are some super outliers in the data. Let us look at the most "asking" and most "answering" users:

```python
idxs = find(X[:,1] > 4000)
for idx in idxs:
    print DBSession.query(User).get(int(X[idx,0])).name, "- Replies to others", X[idx, 1], "Questions", X[idx, 2]
```

```python
idxs = find(X[:,2] > 200)
for idx in idxs:
    print DBSession.query(User).get(int(X[idx,0])).name, "- Replies to others", X[idx, 1], "Questions", X[idx, 2]
```

Given the difference in username styles it seems the most actively replying users are Cisco employees, so we'll drop those and replot:

```python
X = X[find(X[:,1]<2000), :]
scatter(X[:,1], X[:,2])
gca().set_aspect('equal')
xlabel("Replies to others")
ylabel("Questions");
```
